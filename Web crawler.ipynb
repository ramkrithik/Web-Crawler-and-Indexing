```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import WordNetLemmatizer
import requests
import re
from bs4 import BeautifulSoup
```

```python
URL = "https://en.wikipedia.org/wiki/Web_mining"
page=requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')
f=open('web1.txt','w')
f.write(soup.text)
f.close
```

```python
f=open('web1.txt','r')
a=[]
for i in f:
    a.append(i)
f.close()
```

```python
a=[i for i in a if i != '\n']
token=[]
for i in a:
    token.append(word_tokenize(i))
token
```

```python
tokens=[]
for i in token:
    tokens.append([j for j in i if j.isalnum()])
    #for j in i:
     #   if not j.isalnum():
      #      i.remove(j)
tokens
```

```python
stop_words=set(stopwords.words('english'))
a=['.',',','{','}','`','/','``',')','(','[',']',"''",';']
for i in a:
   stop_words.add(i)
```

```python
f=[]
words=[]
s=[]
d={}
for i in tokens:
    f.append([w for w in i if not w in stop_words and w.isalpha()])
    s.append([w for w in i if w in stop_words and w.isalpha()])
    d[' '.join(i)]=[w for w in i if w in stop_words and w.isalpha()]
for i in f:
    for j in i:
        words.append(j.lower())
words=list(set(words))
words=[w for w in words if w.isalnum()]
d
```

```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer() 
p=[]
l=[]
for w in words:
    p.append(ps.stem(w)) 
    l.append(lemmatizer.lemmatize(w))
```

```python
import pandas as pd
data=pd.DataFrame(columns=['term', 'Stemmed Term', 'Lemmetized Word'])
```

```python
data['term']=words
data['Stemmed Term']=p
data['Lemmetized Word']=l
data.head()
```

```python
sentence=[]
stop=[]
tags=[]
for key in d:
    s=list(set(d[key]))
    if(len(s)!=0):
        tag = nltk.pos_tag(s) 
        t=[]
        pos=[]
        for i in tag:
            t.append(i[0])
            pos.append(i[1])
        sentence.append(key)
        stop.append(' , '.join(t))
        tags.append(' , '.join(pos))
```

```python
da=pd.DataFrame(columns=['sentence', 'Stop word', 'POS tag'])
da
```

```python
da['sentence']=sentence
da['Stop word']=stop
da['POS tag']=tags
da.head()
```

```python
fx=open('web2.txt','r')
a=[]
for i in fx:
    a.append(i)
fx.close()
```

```python
a=[i for i in a if i != '\n']
token=[]
for i in a:
    token.append(word_tokenize(i))
token
```

```python
tokens2=[]
for i in token:
    tokens2.append([j for j in i if j.isalnum()])
tokens2
```

```python
f1=[]
for i in tokens2:
    f1.append([w for w in i if not w in stop_words and w.isalpha()])
f_f=[]
f_f1=[]
for i in f:
    for j in i:
        f_f.append(j.lower())
for i in f1:
    for j in i:
        f_f1.append(j.lower())
f=list(set(f_f))
f1=list(set(f_f1))
```

```python
d={}
d1={}
for i in f:
    c=0
    for j in f_f:
        if(i==j):
            c+=1
    d[i]=c
for i in f1:
    c=0
    for j in f_f1:
        if(i==j):
            c+=1
    d1[i]=c
```

```python
bag=pd.DataFrame()
bag
```

```python
bag['DOC']=['doc1','doc2']

for i in f_f:
    if i in f_f1:
        bag[i]=[d[i],d1[i]]
bag.head()
```

```python
boolean=pd.DataFrame()
boolean
```

```python
a1=[]
a2=[]
a3=[]
a=f_f+f_f1
a=list(set(a))
for i in a:
    if i in f_f1 and i in f_f:
        a1.append(i)
        a2.append(1)
        a3.append(1)
    elif i in f_f1 and not i in f_f:
        a1.append(i)
        a2.append(0)
        a3.append(1)
    else:
        a1.append(i)
        a2.append(1)
        a3.append(0) 
boolean['term']=a1
boolean['doc1']=a2
boolean['doc2']=a3
boolean.head()
```
